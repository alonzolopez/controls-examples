```python
import sys
sys.path.append('../')
from simenvs.hcwsim import HCWLTIDynamics
import numpy as np
import plotly.graph_objects as go
import plotly.io as pio
from plotly.offline import download_plotlyjs, init_notebook_mode, plot, iplot
import plotly.figure_factory as ff
import argparse
import plotly.offline as py
# np.random.seed(0)
```

```python
dt = 0.01 # time step

#Initial conditions and parameters given in the problem statement
ca = 1
cb = 0
cc = 0
k1 = 2
k2 = 1

# state space model
X = np.array([ca, cb, cc])
A = np.array([
    [1 - k1*dt, 0, 0],
    [k1*dt, 1 - k2*dt, 0],
    [0, k2*dt, 1]
])
C = np.zeros((3,3))
C[0,0] = 1

# store the state evolution as a list
Xlist = [X]

# run the sim
for i in range(1000):
    Xlist.append(A.dot(Xlist[-1]))
```

```python
# plot
calist = [R[0] for R in Xlist]
cblist = [R[1] for R in Xlist]
cclist = [R[2] for R in Xlist]
dtlist = np.linspace(0,dt*len(calist),num=len(calist))


fig = go.Figure()
fig.add_trace(go.Scatter(y=calist, x=dtlist, mode="lines", name="A"))
fig.add_trace(go.Scatter(y=cblist, x=dtlist, mode="lines", name="B"))
fig.add_trace(go.Scatter(y=cclist, x=dtlist, mode="lines", name="C"))

fig.update_layout(xaxis_title = "Time (sec)", yaxis_title = "Concentration (mol/vol)", title={
        'text': "Chemical Reaction Kinetics",
        'y':0.9,
        'x':0.5,
        'xanchor': 'center',
        'yanchor': 'top'})
fig.show()
```
